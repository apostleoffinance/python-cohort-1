# Data visualization

Data Visualization is the process of representing data graphically to reveal patterns, trends, correlations, and outliers that might not be obvious in raw numbers. It’s essential for data exploration, communication, and decision-making.

### Matplotlib:

A versatile Python library for creating a wide range of plots, from simple lines to complex 3D visuals. It gives you precise control over every element—colors, labels, sizes, and more—making it ideal for custom visualizations.

### Seaborn:

A higher-level library built on Matplotlib, designed for statistical data visualization. It simplifies creating complex plots, integrates easily with Pandas DataFrames, and offers attractive default styles and color palettes.

### Use Case:

Use these tools to analyze datasets like sales records, scientific measurements (e.g., flower dimensions in the Iris dataset), or survey results to identify trends or relationships.

# Matplotlib

Matplotlib is a powerful, flexible Python library for creating static, animated, and interactive visualizations. It’s widely used for data visualization because it allows you to generate a variety of plots—line plots, scatter plots, histograms, bar plots, box plots, and more—while giving you fine-grained control over every aspect of the figure.

### Install/Import Matplotlib

```python
import matplotlib.pyplot as plt 
```

### Basic Setup

```python
student = ['daniel', 'alfred', 'amah', 'divine', 'ruth', 'george']
scores = [80, 90, 70, 60, 85, 74]
```

#### Line Plot

```python
plt.plot(student, scores, color = 'blue')

plt.title('Student Test Score')
plt.xlabel('Students Name')
plt.ylabel('Scores')
plt.show()
```

### Scatter Plot

```python
plt.scatter(student, scores, color = 'red')

plt.title("Student Test Score")
plt.title('Student Test Score')
plt.xlabel('Student Name')
plt.ylabel('Scores')
plt.show()
```

### Bar Plot

```python
plt.bar(student, scores, color = 'green')

#plt.bar(student, scores, color = 'yellow')

plt.title('Student Test Score')
plt.xlabel('Student Name')
plt.ylabel('Scores')
plt.show()
```

### Histogram

```python
plt.hist(scores, color = 'purple')

plt.title('Student Test Score')
plt.xlabel('Student')
plt.ylabel('Score')
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(6, 3))

plt.bar(student, scores, color = 'green')

plt.title('Student Test Score')
plt.xlabel('Student Name')
plt.ylabel('Scores')
plt.show()

```

# Seaborn

Seaborn is a super easy tool in Python that helps you draw pretty pictures (called plots) of your data. It’s like a magic crayon box built on top of Matplotlib, making graphs look nice and simple to create for beginners! Why Use It?
It turns numbers into cool, colorful pictures so you can see patterns fast.

#### Install/Import Seaborn

```python
import seaborn as sns
import pandas as pd
```

### Basic Setup

```python
student = ['daniel', 'alfred', 'amah', 'divine', 'ruth', 'george']
scores = [80, 90, 70, 60, 85, 74]

data = pd.DataFrame({'Students' : student, 'Scores': scores})
print(data)
```

### Line Plot

```python
sns.lineplot(x='Students', y='Scores', data=data, color='blue')
plt.title('Student Test Scores')
plt.show()
```

### Scatter Plot

```python
sns.scatterplot(x = 'Students', y = "Scores", data=data, color = 'red')

plt.title("Student Test Scores")
plt.show()
```

### Bar Plot

```python
sns.barplot(x = 'Students', y = 'Scores', data=data, color='green')

plt.title('Student Test Score')
plt.show()
```

### Histogram

```python
sns.histplot(data=data, x='Scores', color='purple')

plt.title("How scores pile up")
plt.show()
```

# Visualizing Web3 Token Transfer Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/Users/olaoluwatunmise/python-tutorial/Module_2/web3_token_transfers.csv', parse_dates = ['timestamp'])
df['date'] = df['timestamp'].dt.date

# Set Seaborn theme
sns.set(style='whitegrid')
```

```python
df.head()
```

## Token Transfer Volume Distribution

```python
plt.figure(figsize=(10,6))

sns.histplot(df['amount'], bins=50, kde=True, color='skyblue')
plt.title("Distribution of Token Transfer Amounts")
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.xlim(0, 3000)
plt.show()
```

## Daily Total Transfer

```python
daily_volume = df.groupby('date')['amount'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_volume, x='date', y='amount', marker = 'o')
plt.title('Daily Total Token Transfer Volume')
plt.xlabel('Date')
plt.ylabel('Total Amount Transferred')
plt.tight_layout()
plt.show()
```
